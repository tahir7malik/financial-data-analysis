### Data Cleaning

```python
# printing null values of each column

print(f"Null values:")
print(f"[S.No.] = {dataFrame['S.No.'].isnull().sum()}")
print(f"[Name] = {dataFrame['Name'].isnull().sum()}")
print(f"[Mar Cap - Crore] = {dataFrame['Mar Cap - Crore'].isnull().sum()}")
print(f"[Sales Qtr - Crore] = {dataFrame['Sales Qtr - Crore'].isnull().sum()}")
print(f"[Unnamed: 4] = {dataFrame['Unnamed: 4'].isnull().sum()}")
```

```python
# renaming columns

dataFrame.columns = ['S.No.', 'Name of Company', 'Market Cap (in Cr)', 'Quaterly Sales (in Cr)', 'temp-unnamed']
```

```python
# checking our changes in our dataframe

dataFrame
```

```python
# I saw we had 488 rows but serial number starts from 1 & ends with 500
# so, there are some values which may got skipped, lets find them out

max_serial_number = dataFrame['S.No.'].max()
all_numbers = set(range(1, max_serial_number + 1))

# get the set of unique numbers present in the column
present_numbers = set(dataFrame['S.No.'])

# find the numbers that are not present in the column
missing_numbers = all_numbers - present_numbers

# print the missing numbers
print("Numbers not present in ['S.No.'] column:", missing_numbers)
```

```python
# # set display options to show all rows and columns without truncation

# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)
```

```python
dataFrame
```

- temp-unnamed column does not signify important
- Quaterly Sales (in Cr) values has some *NULL* values which got inserted into temp-unnamed column<br><br>
Null values:
1. [S.No.] = 0
2. [Name of Company] = 0
3. [Market Cap (in Cr)] = 9
4. [Quaterly Sales (in Cr)] = 123
5. [temp-unnamed] = 394

```python
# # reset display options to default values

# pd.reset_option('display.max_rows')
# pd.reset_option('display.max_columns')
```

```python
# shifting unnamed-column values into Quateryly Sales (in Cr)

for i in range(len(dataFrame)):
    if pd.isnull(dataFrame['temp-unnamed'][i]):
        pass
    else:
        #dataFrame['Sales Qtr - Crore'][i] = dataFrame['temp-unnamed'][i]
        dataFrame.loc[i, 'Quaterly Sales (in Cr)'] = dataFrame['temp-unnamed'][i]
```

```python
dataFrame
```

- you can see all NaN values in Quaterly Sales (in Cr) are filled with temp-unnamed column values

```python
# deleting temp-unnamed column

del dataFrame['temp-unnamed']
```

```python
del dataFrame['S.No.']
```

```python
dataFrame
```

```python
# lets check quick summary of our dataset

dataFrame.info()
```

```python
# printing updated null values of each column

print(f"Null values:")
print(f"[Name of Company] = {dataFrame['Name of Company'].isnull().sum()}")
print(f"[Market Cap (in Cr)] = {dataFrame['Market Cap (in Cr)'].isnull().sum()}")
print(f"[Quaterly Sales (in Cr)] = {dataFrame['Quaterly Sales (in Cr)'].isnull().sum()}")
```

- Now those left NaN values are both not present in Market Cap & Quaterly Sales
- We will ake out the median of Market Cap & Quaterly Sales individually
- We will update those left NaN values with median

```python
print(f"Median for ['Market Cap (in Cr)']: {dataFrame['Market Cap (in Cr)'].median()}")
print(f"Median for ['Quaterly Sales (in Cr)']: {dataFrame['Quaterly Sales (in Cr)'].median()}")
```

```python
for i in range(len(dataFrame)):
    if pd.isnull(dataFrame['Market Cap (in Cr)'][i]):
        dataFrame.loc[i, 'Market Cap (in Cr)'] = dataFrame['Market Cap (in Cr)'].median()
```

```python
for i in range(len(dataFrame)):
    if pd.isnull(dataFrame['Quaterly Sales (in Cr)'][i]):
        dataFrame.loc[i, 'Quaterly Sales (in Cr)'] = dataFrame['Quaterly Sales (in Cr)'].median()
```

```python
dataFrame.head(10)
```

```python
# checking duplicated valus in Name of Company column

duplicate_values = dataFrame.duplicated(subset=['Name of Company'], keep=False)
print(duplicate_values.sum())
```

- Many of the company names with only caps names have hace spaces between them
- We will strip out the spaces between them

```python
# striping out spaces from only caps company names

for i in range(len(dataFrame)):
    if dataFrame['Name of Company'][i].isupper():
        dataFrame.loc[i, 'Name of Company'] = dataFrame['Name of Company'][i].replace(' ', '')
```

Sales Efficiency Ratio

```python
# this KPI measures the company's ability to generate market capitalization per inr of sales
# higher values indicate better sales efficiency

dataFrame['Sales Efficiency Ratio'] = (dataFrame['Market Cap (in Cr)'] / dataFrame['Quaterly Sales (in Cr)']).round(2)
```

Top 10 companies with highest sales efficiency ratio?

```python
sorted_df = dataFrame.sort_values(by='Sales Efficiency Ratio', ascending=False)
largest_sales_eff_ratio_comp = sorted_df
for i in range(10):
    if sorted_df.iloc[i]['Sales Efficiency Ratio'] == 'inf':
        pass
    print(f"Sales Efficiency Ratio: {sorted_df.iloc[i]['Name of Company']}, {sorted_df.iloc[i]['Sales Efficiency Ratio'].astype(float)}")
```
